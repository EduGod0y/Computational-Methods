O presente relatório mostra a aplicação do método de Euler-Cromer para o problema do pêndulo simples.



```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
 #condições iniciais
w = 1
m = 1
g = 10
l = 10
t = 0
tf = 24
dt = 0.1
theta_list=[]
v_list=[]
t_list=[]
```

###1) Faça gráficos de x por t para os ítens 1, 2, 3 e 4. Calcule o valor do período no caso de pequenos ângulos e comente o que ocorre com o período a medida que o ângulo inicial aumenta.

```python
theta=[(0.1,0),(.5,0),(1.0,0),(3.0,0)]
v=0
ListaV=[]
ListaO=[]

for o,v in theta:
  o_aux = o
  v_list=[]
  t_list=[]
  theta_list=[]
  t=0
  
  #laço euler cromer
  while t<tf:
    o = o + v*dt
    v = v - (w**2)*np.sin(o)*dt
    t = t + dt
    theta_list.append(o)
    v_list.append(v)
    t_list.append(t)
  ListaV.append(v_list)
  ListaO.append(theta_list)
  plt.plot(t_list,theta_list, label=o_aux, linestyle='--')
  plt.ylabel('Posição (rad)')
  plt.xlabel('Tempo (s)')
  plt.title('Gráfico para grandes ângulos')
  plt.legend()

```

Utilizando a aproximação $ sen(\theta) = \theta ; \theta << 1$ percebe-se que o período se comporta como o esperado ($ T = 2 \pi \sqrt{\frac{l}{g}} $) para os ângulos menores e conforme o ângulo inicial aumenta, o erro cresce de maneira que o período também fica maior.
 




```python
for i in range(len(ListaV)):
  plt.plot(ListaO[i],ListaV[i],label=ListaO[i][0])
  plt.legend()
  plt.ylabel('Velocidade (rad/s)')
  plt.xlabel('Posição (rad)')
  plt.title('Espaço de fase para os itens 1,2,3,4')
```

###2) Faça gráficos do espaço de fases, x por v. Compare os comportamentos encontrados para os ítens 1, 2 ,3 e 4 com o encontrado nos ítens 5 e 6.

```python
theta=[(0.1,0),(.5,0),(1.0,0),(3.0,0),(np.pi*3,-.1),(-3*np.pi,.1)]
v=0
ListaV=[]
ListaO=[]
ListaE=[]
for o,v in theta:
  o_aux = o
  v_list=[]
  t_list=[]
  E_list=[]
  theta_list=[]
  t=0
  
  #laço euler cromer
  while t<tf:
    o = o + v*dt
    v = v - (w**2)*np.sin(o)*dt
    t = t + dt
    K=(0.5)*(m*l**2)*(v**2) #cinética
    U=(m*g*l)*(1-np.cos(o))  #potencial
    E = U + K #total
    theta_list.append(o)
    v_list.append(v)
    t_list.append(t)
    E_list.append(E)
  ListaV.append(v_list)
  ListaO.append(theta_list)
  ListaE.append(E_list)
  plt.plot(t_list,theta_list, label=round(o_aux,3), linestyle='--')
  plt.ylabel('Posição (rad)')
  plt.xlabel('Tempo (s)')
  plt.title('Gráfico para grandes ângulos')
  plt.legend()
  

```

```python
for i in range(len(ListaV)):
  plt.plot(ListaO[i],ListaV[i],label=round(ListaO[i][0],3))
  plt.legend()
  plt.ylabel('Velocidade (rad/s)')
  plt.xlabel('Posição (rad)')
  plt.title('Espaço de fase')
```

O espaço de fase mantém um comportamento ocilatório como o do esperado até o item 3, para os itens 4 e 5 o sistema apresenta um comportamento caótico para a movimentação do pêndulo.
